```python
import pandas as pd
import pickle
import json
```

```python
category_mapping = {}
category_mapping['primary_soft_skill'] = 'soft_skills'
category_mapping['primary_hard_skill'] = 'hard_skills'
category_mapping['related_soft_skill'] = 'soft_skills'
category_mapping['related_hard_skill'] = 'hard_skills'
category_mapping['primary_qualification'] = 'unmatched_skills'
category_mapping['primary_transferable_skill'] = 'unmatched_skills'
category_mapping['related_qualification'] = 'unmatched_skills'
category_mapping['related_transferable_skill'] = 'unmatched_skills'
```

```python
base_dir = '/home/gabor/Work/MAGAN/Abodoo/'
data_dir =base_dir+'Data/'
greentech_dir = data_dir+'Greentech/Latest/'
esco_dir = data_dir+'ESCO/'
lightcast_dir = data_dir+'Lightcast/'
output_dir = base_dir+'Data/Output/'
```

```python
df = pd.read_csv(greentech_dir+'Merged_Definitions+Categories_Green_Skills_Taxonomy - Merged_CSV_File - Merged_Definitions+Categories_Green_Skills_Taxonomy - Merged_CSV_File (1).csv', keep_default_na=False)
```

```python
df.head()
```

```python

df.loc[ df['Category'] == '', 'Category'] = 'related_hard_skill'

```

```python
df.Category.unique()
```

```python
greentech_terms = df.Term.to_list()
greentech_category = df.Category.to_list()
greentech_domain =['analytics']*len(greentech_terms)
```

```python
mapped_category = []
for cat in greentech_category:
    try:
        mapped_category.append(category_mapping[cat])
    except:
        mapped_category.append('unmatched_skills')
```

```python
values = [[greentech_domain[i],greentech_category[i],mapped_category[i]] for i in range(len(greentech_terms))]

greentech_terms2cat = dict(zip(greentech_terms,values))
```

Create term category dict from ESCO data combining hard and soft skills

```python
df_esco_hard_skills = pd.read_excel(esco_dir+'New Data.xlsx', sheet_name='Hard Skills',header=None)
```

```python
df_esco_soft_skills = pd.read_excel(esco_dir+'New Data.xlsx', sheet_name='Soft Skills',header=None)
```

```python
df_esco_soft_skills.head()
```

```python
hard_skills_cat =['hard_skills']*len(df_esco_hard_skills)
soft_skills_cat =['soft_skills']*len(df_esco_soft_skills)

```

```python
skill_list = df_esco_hard_skills[df_esco_hard_skills.columns[0]].to_list()+df_esco_soft_skills[df_esco_soft_skills.columns[0]].to_list()
cat_list = hard_skills_cat+soft_skills_cat
domain_list = ['dashboard']*len(skill_list)
values = [[domain_list[i],cat_list[i],cat_list[i]] for i in range(len(skill_list))]
```

```python
esco_terms2cat = dict(zip(skill_list,values))
```

```python
with open(output_dir+'esco_terms2cat.pkl', 'wb') as handle:
    pickle.dump(esco_terms2cat, handle)
```
